## Experiment 2.1 - Spatial features only

```python
df2 = pd.read_csv("../5-Spatial Feature Eng - With Outliers - 2 Ward Features.csv")
df2 = df2.drop_duplicates("ward")
df2 = df2.drop(columns=["log_price"])
print(df2.shape)
df2.head(5)
```

```python
df3 = pd.read_csv("../5-GIS Feature - With Outliers - Table Mountain_and_Beach_and_Ward.csv")
#df3 = pd.concat([df3, df["neighbourhood_cleansed"]], axis=1)
print(df3.shape)
df3.head()
```

```python
df4 = pd.merge(df3, df2, how='left', left_on=['ward'], right_on=['ward'])
#df4.isnull().sum()
df4.head()
```

```python
df2 = pd.read_csv("../5-Spatial data - Wine, Promonade.csv")
df2 = df2[["Promenade", "Wine"]]
df2 = pd.concat([df4, df2], axis=1)

df_spatial_only = df2[["log_price","ward","Low income","Middle-Low Income","Medium Income","High Income","Exceedingly High Income","listing_count","table_mountain_dist","beach_dist","Promenade","Wine"]]
df_spatial_only = pd.concat([df_spatial_only, df_cont_cat_text[["zipcode","latitude","longitude"]]], axis=1)
df_spatial_only.head()


#print(df_spatial_enh["log_price"].corr(df_spatial_enh["Low income"]))

```
